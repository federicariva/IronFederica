```python
#change data type
#change ID from integer to string
#change heart-disease from integer to boolean
strokes["id_str"] = strokes['id'].astype('str')
```

```python
#transformation from str to boolean in one step
strokes['ever_married'] = strokes['ever_married'].replace({ 'No': False, 'Yes': True})
```
