```python
#Create Python function named rentals_may to connect to the Sakila database and 
#return dataframe storing the data from rental table with information for May 2005.

def rentals_may ():
    connection_string = 'mysql+pymysql://root:'+password+'@localhost/sakila'  
    engine = create_engine(connection_string)
    df = pd.read_sql_query("select * from rental where rental_date >= '2005-05-01%%' and rental_date < '2005-06-01';", engine)
    # or pd.read_sql_query('select * from rental where date_format(rental_date, "%%m")='+str(5)+'', engine)
    return df
```

#Write a Python function named check_rentals that given:

- a dataframe with the number of rentals in May by each customer_id
- a dataframe with the number of rentals in June by each customer_id 

return a new dataframe with the following columns:
    - customer_id
    - Rentals_May
    - Rental_June
    - The difference between the previous columns

```python
df_rentbycust_may.info()
```

```python
def check_rentals (df1, df2):
    df3= pd.merge(df1, df2,on='customer_id',how='outer', suffixes = ('_may','_jun'), indicator = 'True')
    #if df3['True'] == ('right only' or 'left only'):
    df3['rental_id_may']=df3['rental_id_may'].fillna(0)
    df3['rental_id_jun']=df3['rental_id_jun'].fillna(0)        
    df3['diff_rent_may_june']= df3['rental_id_jun']-df3['rental_id_may']
    return df3
```

```python
df_rentbycust = check_rentals(df_rentbycust_may,df_rentbycust_jun)
df_rentbycust.info()
df_rentbycust
```
