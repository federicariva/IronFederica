# Challenge 1

```python
df = pd.read_csv('avg-precip-months-seasons.csv')
```

```python
df.plot()
```

```python
df.plot() 
#Set an appropriate xlabel, ylabel, and plot title.
# Add title and axis names
plt.title('average monthly rainfall')
plt.xlabel('months')
plt.ylabel('average rainfall in inches')
```

```python
#Use the linestyle parameter to modify the line style to something other than solid.
df.plot(linestyle = '--')
plt.title('average monthly rainfall')
plt.xlabel('months')
plt.ylabel('average rainfall in inches')
```

```python
#Change the color to something other than the default blue.
df.plot(color = 'lightgreen', linestyle = '--')
plt.title('average monthly rainfall')
plt.xlabel('months')
plt.ylabel('average rainfall in inches')
```

```python
#Add a marker parameter to the ax.plot. What happens when you change the marker in a line plot?
df.plot(color = 'lightgreen', linestyle = '--', marker = "o")
plt.title('average monthly rainfall')
plt.xlabel('months')
plt.ylabel('average rainfall in inches')
```

# Challenge 2

Using the same data you used above, create a bar plot of precipitation data. Once again do the following:

- Read in the .csv file called avg-precip-months-seasons.csv.
- Create a bar plot using ax.bar()
- Set an appropriate xlabel, ylabel, and plot title.
- Use the edgecolor and color parameters to modify the colors of your plot to something other than blue.

```python
df = pd.read_csv('avg-precip-months-seasons.csv')
df.info()
```

```python
#Create a bar plot using ax.bar()
fig, ax = plt.subplots()
ax.bar('months', 'precip', data=df)
plt.show()
```

```python
#Set an appropriate xlabel, ylabel, and plot title.
fig, ax = plt.subplots()
ax.bar('months', 'precip', data=df) # , color='lightblue', edgecolor='darkblue')
ax.set_xlabel ('Months')
ax.set_ylabel ('mean precipitation (inches)')
ax.set_title ('average monthly rainfall')
plt.show()
```

```python
#Use the edgecolor and color parameters to modify the colors of your plot to something other than blue.
fig, ax = plt.subplots()
ax.bar('months', 'precip', data=df, color='lightblue', edgecolor='darkblue')
ax.set_xlabel ('months')
ax.set_ylabel ('mean precipitation (inches)')
ax.set_title ('average monthly rainfall')
plt.show()
```

```python
#Use the edgecolor and color parameters to modify the colors of your plot to something other than blue.
plt.bar(x=df['months'], height=df['precip'], color = 'lightblue', edgecolor = 'darkblue')
plt.xlabel('months')
plt.ylabel('average rainfall (inches)')
plt.title('average monthly precipitation')
```

#Challenge 3: Figure with Two Subplots of Precipitation Data

Above you created two plots:

- a line plot of precipitation data with each point highlighted using a marker.
- a bar plot of precipitation data.

Here, create a single figure that contains two subplots stacked on top of each other.

- The first should be your line plot.
- The second should be your scatter plot.
- For the figure do the following:

-Add an overal title to your figure using plt.suptitle()
-Use plt.tight_layout() to make space between the two plots so that the titles and labels do nor overlap

```python
# Plot the data
fig, (ax1, ax2) = plt.subplots(2,1, figsize=(8, 10))
plt.suptitle("Average precipitaion per month")
ax1.plot('months',
        'precip', data = df,
        color = 'lightgreen', linestyle = '--', marker = "o")

ax1.set(ylabel="mean precipitation (inches)",
       xlabel="months",
       title="average monthly precipitation")

ax2.bar('months',
        'precip', data = df,
       color = 'lightblue', edgecolor = 'darkblue')

ax2.set(ylabel="mean precipitation (inches)",
       xlabel="months",
       title="average monthly precipitation")

plt.tight_layout()
plt.show()
```
