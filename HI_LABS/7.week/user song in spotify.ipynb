#### When the user inputs a song, you are gonna want to retrieve the audio features of that song. How to do it?

- Search the user input using the sp.search() function. This function works similarly to the "search" bar on the spotify app - using Spotify's intelligent search engine. That means that it can handle names of any songs or artists - even certain typos.

- Find the uri of the song that the API gives you back.

- Use sp.audio_features to retrieve the audio features of the song.

```python
# function to extract the uri
def uris_from_playlist(playlist_owner,playlist_id):
    playlist = sp.user_playlist_tracks(playlist_owner,playlist_id)
    playlist_uris=[]
    for item in playlist['items']:
        playlist_uris.append(item['track']['uri'])
    return playlist_uris
```

```python
def uris_from_playlist(playlist_id):
    playlist = sp.playlist_tracks(playlist_id)
    playlist_uris=[]
    for item in playlist['items']:
        playlist_uris.append(item['track']['uri'])
    return playlist_uris
```
